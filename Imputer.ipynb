```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
data=pd.read_csv("Missing-data.csv")
data
```

```python
X=data.iloc[:,0:3]
Y=data.iloc[:,3]
X
```

```python
Y=pd.DataFrame(Y,columns =['Purchased'])
Y
```

```python
imputer=SimpleImputer(missing_values = np.nan,strategy ='mean')
updated=imputer.fit_transform(X.iloc[:,1:])
updated
```

```python
X.iloc[:,1:]=updated
X
```
