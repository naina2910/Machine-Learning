```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
missing_values=['*','**']
df=pd.read_csv('share.csv',na_values=missing_values)
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Average Deaths per Year'].fillna(0,inplace=True)
```

```python
df['Average Deaths per Year'].drop([0],axis=0,inplace=True)
```

```python
df.shape
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

```python
df['Met Objective of 45.5? (1)'].fillna("Not Avilable",inplace=True)
```

```python
df['Age-Adjusted Death Rate'].fillna(df['Age-Adjusted Death Rate'].mean(),inplace=True)
```

```python
df['Lower 95% Confidence Interval for Death Rate'].fillna(df['Lower 95% Confidence Interval for Death Rate'].mean(),inplace=True)
```

```python
df['Lower 95% Confidence Interval for Trend'].replace(np.NaN,df['Lower 95% Confidence Interval for Trend'].mean(),inplace=True)
```

```python
df.drop('Average Deaths per Year',axis=1)
```

```python
df.dropna(thresh=3,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(how='all',inplace=True)
```
